Miremos el código postal. Es una variable con entradas erroneas.
Hay valores como '', '-'  y '28' que deberían ser considerados como NA.
Así mismo también debería ser NA todos los que no compiencen por 28, ya que estamos con códigos postales de Madrid

El código postal 28002, 28004 y 28051 tienen entradas repetidas. 
Por ejemplo las entradas 28002\n20882 deberían ir dnetro de 28002

El codigo 2804 debería ser 28004, 2805 deberia ser 28005 y 2815 juncto con 2815 debería ser 28015

Limpia los datos de la columna Codigo Postal

```r
## En dos pasos. Con un guión
guiones <- airbnb$CodigoPostal == '-'
airbnb$CodigoPostal[guiones] <- NA

## Forma rápida
# Los Vacíos <- NA
airbnb$CodigoPostal[airbnb$CodigoPostal == ''] <- NA
# El '28' <- NA
airbnb$CodigoPostal[airbnb$CodigoPostal == '28'] <- NA
# No empiecen por 28
airbnb$CodigoPostal[!grepl('28', airbnb$CodigoPostal)] <- NA

# Entradas repetidas
airbnb$CodigoPostal[airbnb$CodigoPostal == '28002\n28002'] <- NA
airbnb$CodigoPostal[airbnb$CodigoPostal == '28004\n28004'] <- NA
airbnb$CodigoPostal[airbnb$CodigoPostal == '28051\n28051'] <- NA

# Añadir 0
airbnb$CodigoPostal[airbnb$CodigoPostal == '2804'] <- '28004'
airbnb$CodigoPostal[airbnb$CodigoPostal == '2805'] <- '28005'
airbnb$CodigoPostal[airbnb$CodigoPostal == '2815'] <- '28015'

# Otros
airbnb$CodigoPostal[airbnb$CodigoPostal == 'Madrid 28004'] <- '28004'
airbnb$CodigoPostal[airbnb$CodigoPostal == '280013'] <- '28013'
```

Muestra el diagrama de cajas del precio para cada uno de los diferentes Tipos de Alquiler

```r
library("ggplot2")
options(repr.plot.width=20, repr.plot.height=20, repr.plot.res = 300)

con_precio <- airbnb[!is.na(airbnb$Precio), ]
ggplot(data=con_precio, aes(x=TipoAlquiler, y=Precio, color=TipoAlquiler))+
   geom_boxplot(na.rm = FALSE) +
   scale_color_discrete(name="Tipo Alquiler")
```

Cual es el precio medio de alquiler de cada uno, la diferencia que hay ¿es estadísticamente significativa?
¿Con que test lo comprobarías?

```r
library(dplyr)
```

```r
con_precio |>
    group_by(TipoAlquiler) |>
    summarise(PrecioMedio = mean(Precio))

# Comprobar si son gausianas los precios
paste("Las muestras para tipo de alquiler \"Entire home/apt\" no parecen seguir una gausiana. pvalor:", 
      shapiro.test(con_precio[con_precio$TipoAlquiler == 'Entire home/apt', 'Precio'][1:5000])$p.value)

paste("Las muestras para tipo de alquiler \"Shared room\" no parecen seguir una gausiana. pvalor:", 
      shapiro.test(con_precio[con_precio$TipoAlquiler == 'Shared room', 'Precio'][1:5000])$p.value)

paste("Las muestras para tipo de alquiler \"Private room\" no parecen seguir una gausiana. pvalor:", 
      shapiro.test(con_precio[con_precio$TipoAlquiler == 'Private room', 'Precio'][1:5000])$p.value)

# Como NO son gausianas se tiene que usar el test de Kruskal para compara las medias

kruskal.test(Precio ~ TipoAlquiler, data=con_precio)
paste("Realizando el test Kruskal, porque no sigue una distribución normal, parece ser que la diferencia es significativa. p-valor bajo")
```

Esta claro que las medias de metros cuadrados de cada uno de estos 5 barrios parecen ser diferentes, pero ¿son estadísticamente diferentes?
¿Que test habría que usar para comprobarlo?

```r
df_5_barrios <- filter(con_metros, Barrio %in% lista_barrios)

# Comprobar si son gausianas los precios
paste("El barrio de Goya no parece disponer de las suficientes muestras para realizar el test de shapiro. Numero de muestars: ",
      nrow(df_5_barrios[df_5_barrios$Barrio == 'Goya', ]))
paste("Las muestras para el barrio \"Justicia\" parecen seguir una gausiana. pvalor:", 
      shapiro.test(df_5_barrios[df_5_barrios$Barrio == 'Justicia', 'MetrosCuadrados'])$p.value)
paste("Las muestras para el barrio \"Recoletos\" NO parecen seguir una gausiana. pvalor:", 
      shapiro.test(df_5_barrios[df_5_barrios$Barrio == 'Recoletos', 'MetrosCuadrados'])$p.value)
paste("Las muestras para el barrio \"Trafalgar\" parecen seguir una gausiana. pvalor:", 
      shapiro.test(df_5_barrios[df_5_barrios$Barrio == 'Trafalgar', 'MetrosCuadrados'][1:5000])$p.value)
paste("Las muestras para el barrio \"Sol\" NO parecen seguir una gausiana. pvalor:", 
      shapiro.test(df_5_barrios[df_5_barrios$Barrio == 'Sol', 'MetrosCuadrados'][1:5000])$p.value)

# Como me salen unos barrios que si son gausianos y otros dos paso ambos tests ANOVA y Kruskal
summary(aov(MetrosCuadrados ~ Barrio, data=df_5_barrios))
kruskal.test(MetrosCuadrados ~ Barrio, data=df_5_barrios)

paste("Ambos test tienen un p-value alto con lo cual podemos concluir que NO son estadísticamente diferentes")
```

Calcular un modelo lineal que combine alguna de estas variables:
* NumBanyos
* NumDormitorios
* MaxOcupantes
* MetrosCuadrados

```r
barrio_sol<-subset(airbnb_entire,Barrio=="Sol")
```

Primero calculamos la correlación para ver como se relacionan estas variables entre sí.

```r
# Limpiamos los datos
df_sol_limpio <- barrio_sol[
    !is.na(barrio_sol$NumBanyos) & 
    !is.na(barrio_sol$NumDormitorios) & 
    !is.na(barrio_sol$MaxOcupantes) & 
    !is.na(barrio_sol$MetrosCuadrados), ]

# Se calcula la matriz de correlaciones entre las variables
cor(
    cbind(
        df_sol_limpio$NumDormitorios,
        df_sol_limpio$MetrosCuadrados,
        df_sol_limpio$NumBanyos,
        df_sol_limpio$MaxOcupantes
    )
)
```

Se observa que la correlación entre el número de dormitorios y los metros cuadrados es sorprendentemente baja.
¿Son de fiar esos números?

Mediante un histograma o curvas de densidad podemos descartar números que notienen sentido en el dataframe barrio_sol,
para tener una matriz de correlación que tenga mayor sentido.


```r
ggplot(df_sol_limpio, aes(x=MetrosCuadrados)) +
   geom_density()

# Voy a descartar los pisos con menos de 20m2
df_sol_mas_20_metros <- subset(df_sol_limpio,df_sol_limpio$MetrosCuadrados > 20)

cor(
    cbind(
        df_sol_mas_20_metros$NumDormitorios,
        df_sol_mas_20_metros$MetrosCuadrados,
        df_sol_mas_20_metros$NumBanyos,
        df_sol_mas_20_metros$MaxOcupantes
    )
)
```

Una vez que hayamos filtrado los datos correspondientes calcular el valor o la combinación de valores que mejor nos permite obtener el precio de un inmueble.

```r
model<-lm(data=df_sol_mas_20_metros, formula = Precio ~ NumDormitorios + MetrosCuadrados + NumBanyos + MaxOcupantes + NumCamas)
```

¿Que variable es más fiable para conocer el precio de un inmueble, el número de habitaciones o los metros cuadrados?

```r
summary(model)
paste("La variable más fiable para calcular el precio es \"NumDormitorios\", es decir el número de habitaciones")
```

Responde con su correspondiente margen de error del 95%, ¿cuantos euros incrementa el precio del alquiler por cada metro cuadrado extra del piso?

```r
model_precio_metro <- lm(data=df_sol_mas_20_metros, formula = Precio ~ MetrosCuadrados)
summary(model_precio_metro)
paste("El precio por metro cuadrado se incrementa en: ", model_precio_metro$coefficients[2])
confint(model_precio_metro)
```

Responde con su correspondiente margen de error del 95%, ¿cuantos euros incrementa el precio del alquiler por cada habitación?

```r
model_precio_habitacion <- lm(data=df_sol_mas_20_metros, formula = Precio ~ NumDormitorios)
summary(model_precio_habitacion)
paste("El precio por habitación se incrementa como media en: ",model_precio_habitacion$coefficients[2])
confint(model_precio_habitacion)
```
